### load data

```python
import pandas as pd
```

```python
example_df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
example_df
```

### ดูการกระจายของข้อมูลในตารางๆแบบ คร่าวๆ

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=[8,8] #กำหนดขนาดของ figure
```

```python
pd.plotting.scatter_matrix(example_df); # กราฟแสดงความสัมพันธ์
```

จากกราฟมีความสัมพันธ์กัน ต้องการลดจำนวนข้อมูลจึงทำการทำ PCA

### เริ่มทำ PCA

sklearn -> scikit-learn เป็น package ที่รวบรวม function การทำ Data Science - Machine Learning - Data Mining เอาไว้ให้เราใช้แบบไม่ต้องเขียนเอง

import

```python
from sklearn.decomposition import PCA
```

define

```python
pca = PCA() # ประกาศตัวแปร
```

fit-transform
ทรานฟอร์มตัวแปรทั้ง 4 ตัว

```python
thisdata = example_df.iloc[:,:-1] #
thisdata
```

```python
new_axis = pca.fit_transform(thisdata) #การทำ PCA
```

```python
new_axis.shape
```

```python
new_axis
```

```python
PCAdf = pd.DataFrame.from_records(new_axis, columns = ['PCA1','PCA2','PCA3','PCA4']) #ได้ตัวแปรทั้งหมด 4 ตัวแปร แล้วตั้งชื่อ
PCAdf
```

เกณฑ์การเลือกว่าเราจะเลือกอะไรใน PCA 2 แกน [variance (80ขึ้น) , ไอเทม value มากกว่า1 ]

```python
pca.explained_variance_ratio_ #
```
